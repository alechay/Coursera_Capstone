```python
import geopandas

# read geojson file
geo = geopandas.read_file('newdata/Neighborhood_Clusters.geojson')
# get centroid of geometries
geo['centroid'] = geo.centroid
# get coords of column
geo['coords'] = geo['centroid'].apply(lambda x: list(x.coords)[0])
# separate coords into lat and long
geo[['lon','lat']] = pd.DataFrame(geo.coords.tolist(), index= geo.index)
# display head again
geo.head()
```

```python
# define function to get nearby food

def getNearbyVenues(names, neighborhoods, latitudes, longitudes, radius=250, categoryId='4d4b7105d754a06374d81259'):

    venues_list = []
    for name, nbh, lat, lng in zip(names, neighborhoods, latitudes, longitudes):

        #create API request
        url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            VERSION,
            lat,
            lng,
            radius,
            LIMIT,
            categoryId
        )

        # make GET request
        results = requests.get(url).json()['response']['venues']

        # return only relevant information for each nearby venue
        venues_list.append([(
            name,
            nbh,
            lat, 
            lng, 
            v['name'],
            v['id'],
            v['location']['lat'], 
            v['location']['lng'],  
            v['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Cluster',
                             'Neighborhood', 
                             'Neighborhood Latitude', 
                             'Neighborhood Longitude', 
                             'Venue',
                             'Venue ID',
                             'Venue Latitude', 
                             'Venue Longitude', 
                             'Venue Category']
    
    return(nearby_venues)
```

### Read in additional data and select features

```python
# read in all the data
crime=pd.read_csv('newdata/safety_cl17.csv')
income=pd.read_csv('newdata/income_cl17.csv')
employment=pd.read_csv('newdata/employment_cl17.csv')
population=pd.read_csv('newdata/population_cl17.csv')
housing=pd.read_csv('newdata/housing_cl17.csv')
```

```python
# take 2016 data from crime (data is year by year)
c=crime.loc[crime['timeframe']==2016][['cluster2017_nf','Crimes_pt1_property_per1000', 'Crimes_pt1_violent_per1000']]
# take 2012-2016 data from income (data is estimated over this period)
i=income.loc[(income['timeframe']=='2012-16') & (income['AvgFamilyIncAdj']!='X')][['cluster2017_nf', 'AvgFamilyIncAdj']]
# take 2012-2016 data from employment (data is estimated over this period)
e=employment.loc[employment['timeframe']=='2012-16'][['cluster2017_nf', 'PctUnemployed']]
# take 2012-2016 data from population (data is estimated over this period)
p=population.loc[population['timeframe']=='2012-16'][['cluster2017_nf', 'TotPop']]
# take 2010 data from housing (data not available for 2012-2016)
h=housing.loc[(housing['timeframe']=='2010') & (housing['mprice_sf']!='X')][['cluster2017_nf','mprice_sf']]
```

### Join with the venue per cluster data

```python
# set index to cluster to concat the df's
data=[c,i,e,p,h]
for x in data:
    x.set_index('cluster2017_nf', inplace=True)
```

```python
# concat the df's then join with venue data
full=num_venues_per_cluster.join(pd.concat([c, i, e, p, h], axis=1)).reset_index()
```

```python
# strip 'cluster' from col values and set cluster # to integer
full['Cluster']=full['index'].str.strip('Cluster ').astype(int)
# set cluster # as index and sort
full=full.drop(['index', 'Neighborhood'], axis=1).set_index('Cluster').sort_index()
# drop clusters 40 and 42--they have incomplete data. And convert all dtypes to numeric
full=full.drop([40, 42]).apply(pd.to_numeric)
```

```python
# show head of data
full.head()
```

### Cluster Neighborhoods based on all this data

```python
from sklearn import preprocessing

# scale data
scaled=preprocessing.scale(full.to_numpy())
```

```python
from sklearn.cluster import KMeans

# do kmeans clustering of neighborhoods
kclusters=3
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(scaled)
```

```python
# select columns from dc venues
cluster_geo=dc_venues[['Cluster', 'Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude']].drop_duplicates()
# # strip 'cluster' from col values and set cluster # to integer
cluster_geo['Cluster']=cluster_geo['Cluster'].str.strip('Cluster ').astype(int)
# set cluster # as index and sort
cluster_geo=cluster_geo.set_index('Cluster').sort_index()
# drop clusters 40 and 42--they have incomplete data.
cluster_geo.drop([40, 42], inplace=True)
# insert cluster labels
cluster_geo.insert(0, 'Cluster Labels', kmeans.labels_)
# show df
cluster_geo.head()
```

### Map of clustered neighborhoods

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(cluster_geo['Neighborhood Latitude'], cluster_geo['Neighborhood Longitude'], cluster_geo['Neighborhood'], cluster_geo['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

## 2. Which features (number of restaurants, crime, income, employment, population) are most correlated with housing prices?

```python
# Separate into features and labels
X = scaled[:, :-1] # for all but last column
y = scaled[:, -1] # for last column
```

```python
# get array of colors for easier visualization
colors_array = cm.rainbow(np.linspace(0, 1, X.shape[1]))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
# check if features are normally distributed
for col in range(X.shape[1]):
    plt.figure()
    plt.hist(X[:,col], color=rainbow[col])
    plt.title(full.columns[col])
```

```python
# Check if labels are normally distributed
plt.hist(y)
plt.title('Housing price');
```

```python
# get data to examine correlations
df=examine_clusters.drop(columns=['Cluster', 'Cluster Labels', 'Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude'])

#Using Pearson Correlation
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

Violent crimes, average family income, and percentage unemployed are most correlated with mean housing price. Venue density does not predict housing price well (0.4).

```python
#Correlation with output variable
cor_target = abs(cor['mprice_sf'])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.5]
relevant_features
```

## 3. Can we create a model, based on those features, that predicts housing prices? Which type of model makes the best predictions?

```python
from sklearn.model_selection import train_test_split

# train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# instantiate linear regression
reg = LinearRegression()
# fit to training data
reg.fit(X_train, y_train)
# make predictions
y_pred = reg.predict(X_test)
# score predictions
print('Mean squared error is:', mean_squared_error(y_test, y_pred))
```

### Random Forest (ensemble technique)

```python
from sklearn.ensemble import RandomForestRegressor

# instantiate random forest
rf= RandomForestRegressor(max_depth=2, random_state=0)
# fit to training data
rf.fit(X_train, y_train)
# make predictions
y_pred = rf.predict(X_test)
# score predictions
print('Mean squared error is:', mean_squared_error(y_test, y_pred))
```

### Gradient boosting (ensemble technique)

```python
from sklearn.ensemble import GradientBoostingRegressor

# instantiate gradient boosting
gbr = GradientBoostingRegressor(random_state=0)
# fit to training data
gbr.fit(X_train, y_train)
# make predictions
y_pred = gbr.predict(X_test)
# score predictions
print('Mean squared error is:', mean_squared_error(y_test, y_pred))
```
